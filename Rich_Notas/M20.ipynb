$$ 
     \cos (x) = \sum_{n=0}^{\infty} -1^{n}\frac{x^{2n}}{(2n)!} = 1- \frac{x^{2}}{2} +\frac{x^{4}}{24}+...
$$

```python
def cos(x,n_max=5 ):
    y = 0
    for n in range(n_max):
        y = y + ((-1)**(n))*((x**(2*n))/mt.factorial(2*n)) #y = y + sm es lo mismo que y += sm 
    return y
```
