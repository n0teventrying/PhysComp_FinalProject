```python
def mi_cos(x,n_max=15):
    y = 0
    x = x%(2*pi)
    for n in range(n_max):
        y += ((-1)**(n))*((x**(2*n))/mt.factorial(2*n))
    return y
```

```python
%%timeit
mi_cos(pi/4)
```
