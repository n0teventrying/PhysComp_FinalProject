```python
from pylab import *
```

```python
x = linspace(0,2*pi,100)
y = cos(x**2)
```

```python
pi
```

```python
title('Mi gráfica super acá')
#Pues lit lo que dice padrino, es pa que tenga titulo tu gráfica no
#como yo, que ya voy en decimo y ni las luces del titulo xD
plot(x,y,label=r'$\cos(x^3)$')
#Plot es grafica, toma los puntos que definimos como x, y los de y
#Ya aquí para label es un poco más complicado, depende de que le quieras
#meter como label a la gráfica, texto, números.

xlabel('x') , ylabel('y')
#Pues eso no, la label del eje x, y la del eje y
xlim([0,max(x)])
#Aquí le pones un límite a x, porque sino creo que se sigue la gráfica
grid(), legend()
#grid, es como esa madre que se ponen los chefs para que no se les caiga
#cabello, entonces su analogo aqui es como el fondo de tela cuadrada,
#algo asi como una red,bueno lo alcanzas a ver si o no meco??
#legend, esa madre es para que se vean las leyendas
show()
#seriously pinche codos negros, pues es show wey
```
