```python
def Taylor2V20MODEADA(F, Fy, u0, t,omega):
    n = len(u0)
    u = zeros((len(t), n))
    u[0] = u0
    h = t[1] - t[0]
    for i in range(0, len(t) - 1):
        k1 = F(u[i], t[i],omega)
        k2 = dot(Fy(u[i], t[i],omega), k1)
        u[i+1] = u[i] + h * k1 + 0.5 * h**2 * k2
    return u

def F(u, t,omega):
    x, y = u
    return array([y, -0.1 * y - 4 * sin(x) + 0.5 * sin(omega * t)])

def Fy(u, t,omega):
    x, y = u
    return array([[0, 1], [-4 * cos(x), -0.1]])

u0 = [0, 1]
t = linspace(0, 10, 100)

figure(figsize=(10, 6))

for omega in arange(1, 3, 0.1):
    X = Taylor2V20MODEADA(F,Fy, u0, t, omega)
    plot(t, X[:, 0], label=f'omega={omega:.1f}')
    
legend(bbox_to_anchor=(1.05, 1.0), loc='upper left')
title('Soluciones para distintos valores de omega con Taylor2')
show()
```
