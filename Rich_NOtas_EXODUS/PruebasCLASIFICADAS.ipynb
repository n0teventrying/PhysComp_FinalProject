```python
def deriv_h5(x,f):
    f_prima = zeros(len(f))
    h = x[1] - x[0]
    doceh = 12*h
    for i in range(2,len(f)-2):
        f_prima[i] = ( -f[i+2]+ 8*(f[i+1]-f[i-1]) +f[i-2] )/(doceh)
    c0,c1,c2,c3,c4 = -25/12,4,-3,4/3,-0.25
    # calcular el primer punto
    f_prima[0] = ( c0*f[0] + c1*f[1] + c2*f[2] + c3*f[3] + c4*f[4] )/h
    # calcular el segundo punto
    f_prima[1] = ( c0*f[1] + c1*f[2] + c2*f[3] + c3*f[4] + c4*f[5] )/h
    # calcular el ultimo punto o el punto N
    f_prima[-1] = -( c0*f[-1] + c1*f[-2] + c2*f[-3] + c3*f[-4] + c4*f[-5] )/h
    # calcular el penultimo punto o el N-1
    f_prima[-2] = -( c0*f[-2] + c1*f[-3] + c2*f[-4] + c3*f[-5] + c4*f[-6] )/h
    return f_prima
```

```python
x = linspace(0,pi,100)
f = sin(x)
#derivada analtica
dy = cos(x)
dy_h2 = deriv_h5(x,f)
```

```python
figure(dpi=150)
plot(x,dy,label="f' analitica ")
plot(x,dy_h2,label="f' numerica")
legend()
show()
```

```python
import matplotlib.pyplot as plt

# Example array of points
points = [(1, 2), (3, 4), (5, 6), (7, 8)]

# Extract x and y coordinates from the array
x_coords = [point[0] for point in points]
y_coords = [point[1] for point in points]

# Compute the integral of the function defined by the points using the trapezoidal rule
integrated_y_coords = [0] * len(y_coords)
for i in range(1, len(y_coords)):
    integrated_y_coords[i] = integrated_y_coords[i-1] + (y_coords[i-1] + y_coords[i]) * (x_coords[i] - x_coords[i-1]) / 2

# Create a plot of the primitive function
plt.plot(x_coords, integrated_y_coords)

# Display the plot
plt.show()
```
