$$
f(x_{0}+h) = f(x_{0}) +hf'(x_{0})+ \frac{h^2}{2!}f''(x_{0})+\frac{h^3}{3!}f'''(x_{0})+ \frac{h^4}{4!}f^{iv}(x_{0}) + \mathcal{O}(h^5)
$$
$$
f(x_{0}+h) = f(x_{0}) -hf'(x_{0})+ \frac{h^2}{2!}f''(x_{0})-\frac{h^3}{3!}f'''(x_{0})+ \frac{h^4}{4!}f^{iv}(x_{0}) + \mathcal{O}(h^5)

$$
f(x_{0}-2h) = f(x_{0}) -2hf'(x_{0})+ 2h^2f''(x_{0})-\frac{4h^3}{3}f'''(x_{0})+ \frac{2h^4}{3}f^{iv}(x_{0}) + \mathcal{O}(h^5)

Si lo vemos como un sistema de ecuaciones, entonces:

$$
 hx_{1} + \frac{h^2}{2!}x_{2} + \frac{h^3}{3!}x_{3} +\frac{h^4}{4!} x_{4} = a-w 
 $$ 
 $$
-hx_{1}+\frac{h^2}{2!}x_{2} -\frac{h^3}{3!}x_{3} + \frac{h^4}{4!}x_{4} = b-w $$
$$
2h x_{1} + 2h^2x_{2} + \frac{4h^3}{3}x_{3} + \frac{2h^4}{3}x_{4}= c -w 
$$ 
$$
-2h x_{1} +2h^2x_{2} -\frac{4h^3}{3}x_{3} + \frac{2h^4}{3}x_{4}= d-w    
$$

donde le asocie los valores $a, b, c, d, w$ a $f(x_{0}+h), f(x_{0}-h), f(x_{0}+2h), f(x_{0}-2h),f(x_{0})$ respectivamente porque los conocemos.

$$

\right)

Entonces encontramos que para la derivada:

$$f'(x_{0}) =\frac{8f(x_{0}+h)-8f(x_{0}-h)-f(x_{0}+2h)+f(x_{0}-2h) }{12h} $$ 

Mhhh, para programar para los extremos, vamos a usar puntos dentro de la malla, y enfrente de nuestro punto de interes,es decir, $x+h,x+2h,x+3h,x+4h$.

### Comparación de la primera y segunda derivada de f(x), g(x) númericas con aproximación de segundo orden 2 con sus contrapartes analitícas así como su error para cada una.
Bueno, recordando el algoritmo, obviamente lo copie de la notebook de la clase.

```python
def deriv_h2(x,f):
    #Esto funciona si la malla es uniforme
    f_prima = zeros_like(f) # es un array con puros ceros LOL de ña longitud de f
    h = (x[1]-x[0])
    for i in range(len(f)-1):
        f_prima[i] = (f[i+1]-f[i])
    f_prima[-1] = (f[-1]-f[-2])
    return f_prima/h
```

### Comencemos con f(x)

```python
w = linspace(-5,5,10000)
f = e**(-w**2)*sin(10*w)
f_prima = -e**(-w**2)*(2*w*sin(10*w)-10*cos(10*w))
f_prima_ord2 = deriv_h2(w,f)
plot(w,f_prima,'.',label='f\' Analitica')
plot(w,f_prima_ord2,label=' f\' Orden 2')
title('Derivada númerica de orden 2 comparada con analitíca')
legend()
show()
```

### Error f' de orden 2

```python
#definimod una función de error, que segun yo escupe la diferencia entre los valores, pero me grafica esa cosa rara xd
def Error(f_1,f_2):
    O = zeros(len(f_1)) 
    if len(f_1)==len(f_2):
        for i in range(len(f_1)):
            O[i] = abs(f_1[i]-f_2[i])
    else:
        print('no es del mismo tamaño tus funciones bro')label='primitiva de aceleración_x'
    return O
    
    
```
