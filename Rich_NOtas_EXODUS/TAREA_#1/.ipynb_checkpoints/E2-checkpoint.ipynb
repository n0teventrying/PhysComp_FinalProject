```python
#Bueno, antes de empezar a programar solo estaba extendiendo el 
#procedimiento que vimos en clase pero ahora que lo estoy programando
#bueno, se tiene que sumar todo xD

def derivH1(x,f):
    f_prima = zeros(len(f)) # pues bro, en algún lugar tenemos que 
    #guardar los datos que vamos a generar.
    h = x[1]-x[0] #definimos nuestro 
#como se ve en la central, los puntos que no podemos calcular son
#los dos primeros y los dos ultimos, entonces por eso el rango del
#for empieza en 2 y termina 2 puntos antes de la len(f)
    for i in range(2,len(f)-2):
        f_prima[i] = (8*f[i+1]-8*f[i-1]-f[i+2]+f[i-2])/(12*h)
#ahora, necesitamos llenar los puntos que nos faltan de f_prima
#f[0],f[1],f[len(f)-1],f[len(f)].
#Y, me dio error xD, al parecer el último termino se llama con
#len(f)-1, y por ende el penúltimo con len(f)-2, entonces
    
    f_prima[0] = ( -(25/12)*f[0] + 4*f[1] -3*f[2] + (4/3)*f[3] - (1/4)*f[4] )/h
    f_prima[1] = ( -(25/12)*f[1] + 4*f[2] -3*f[3] + (4/3)*f[4] - (1/4)*f[5] )/h
    f_prima[len(f)-2] = -( -(25/12)*f[len(f)-2] + 4*f[len(f)-3] -3*f[len(f)-4] + (4/3)*f[len(f)-5] - (1/4)*f[len(f)-6] )/h
    f_prima[len(f)-1] = -( -(25/12)*f[len(f)-1] + 4*f[len(f)-2] -3*f[len(f)-3] + (4/3)*f[len(f)-4] - (1/4)*f[len(f)-5] )/h   
    return f_prima

#Al parecer jala, solo que no sé porque le tengo que meter signos
#menos en las ultimas dos primas xD, ahorita vemos, ah ya, es porque 
#vamos en sentido contrario y la h escupe un signo menos ahora.
#sencillo.
    
```

```python
#Ahora supongo es algo más sencillo, porque solo nos debemos
#preocupar por 2 casos, cuando la malla es impar o par, entonces.
#Epa, no es tan sencillo, lo sé no me compilo xD
#Bueno, si bien es cierto que tenemos dos casos, en los puntos finales
#cambia la cosa

def intH1(x,f):
    integral = 0
    h = x[1] - x[0]
    Wasabi = len(f)
    if mod(Wasabi,2) == 0: #Par
        for i in range(1,len(f)-2,2):
            integral += h*((19/15)*f[i]+(17/45)*f[i+1]+(17/45)*f[i-1]-(1/90)*f[i+2]-(1/90)*f[i-2])
        
        integral += h*((251/720)*f[0] + (323/360)*f[1] -(11/30)*f[2] + (53/360)*f[3] - (19/720)*f[4])
        
        integral += (h*((251/720)*f[-1] + (323/360)*f[-2] -(11/30)*f[-3] + (53/360)*f[-4] - (19/720)*f[-5]))
    else:
        for i in range(1,len(f)-2,2):
            integral += h*((19/15)*f[i]+(17/45)*f[i+1]+(17/45)*f[i-1]-(1/90)*f[i+2]-(1/90)*f[i-2])
        
        integral += h*((251/720)*f[0] + (323/360)*f[1] -(11/30)*f[2] + (53/360)*f[3] - (19/720)*f[4])
        
        integral += (h*((251/720)*f[-1] + (323/360)*f[-2] -(11/30)*f[-3] + (53/360)*f[-4] - (19/720)*f[-5]))
            
        integral += (h*((251/720)*f[-2] + (323/360)*f[-3] -(11/30)*f[-4] + (53/360)*f[-5] - (19/720)*f[-6]))  
    return integral

```
