3.Calcule la derivada analítica de las funciones $f(x)=e^{-x^{2}}\sin(10x)$ con $x \in [-5,5]$, y $g(x)=-\frac{2}{x^6}+\frac{1}{x^{12}}$con $x \in [0.5,5]$. Realice las gráficas de las funciones y sus primeras dos derivadas.\
Con los algoritmos vistos en clase (los que desprecian términos mayores a $\mathcal{O}(h^2)$), encuentre la derivada númerica de dichas funciones con error menor a $10^{-14}$ para cualquier punto y compare con las derivadas analíticas.\
También realice una comparación utilizando la función derivada del problema anteior. Es decir, haga una gráfica en la que se muestre $|f'_{analítica}(x)-f'_{h^2}|$ y $|f'_{analítica}(x)-f'_{h^4}|$, así como para las segundas derivadas, y muestre que el error es menor a $10^{-4}$ en todo el intervalo con su valor de h respectivo.

Okay, entonces calculando las respectivas primera y segunda derivada de dichas funciones tenemos que:
$$
f'(x)=-e^{-x^2}(2x\sin(10x)-10\cos(10x))
$$\
$$
f''(x)=e^{-x^2}((4x^2-102)\sin(10x)-40x\cos(10x))
$$\
$$
g'(x)= 12(x^{-7}-x^{-13})
$$\
$$
g''(x)=-84x^{-8}+156x^{-14}
$$

```python
x = linspace(-5,5,100)
f = e**(-x**2)*sin(10*x)
f_prima = -e**(-x**2)*(2*x*sin(10*x)-10*cos(10*x))
f_biprima = e**(-x**2)*((4*x**2-102)*sin(10*x)-40*x*cos(10*x))
plot(x,f)
plot(x,f_prima)
plot(x,f_biprima)
show()
```
