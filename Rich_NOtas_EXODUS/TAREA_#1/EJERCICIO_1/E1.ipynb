```python
#Bueno, en este caso tenemos tres columnas, la x,y,z, ¿qué hace cada una de estas?
#buena pregunta, la x no es relevante para nuestro movimiento, la z menos, porque se deberia ver como unos picos, 
#aún así, el ejercicio dice que recupere todos, es decir 3 columnas, ya a la hora de graficar es distinto.

from pylab import*
import csv 

#aquí solo importe la libreria de pylab y csv.
```

```python
datos = []
#declaro a datos como una lista
with open('E1_DATOS.csv') as csvfile:
    datos_leidos = csv.reader(csvfile,delimiter=',')
#aquí solo es eso, lo hago un objeto leible al E1, y le digo que tipo de cosa diferencia un dato de otro, la coma
    for renglon in datos_leidos: 
        datos.append(renglon)
N_datos = len(datos[1:])
datos_numericos = zeros((4,N_datos))

#Aquí estuve jugando un poco cambiando los argumentos, pero solo 
#descubri que es lit, solo prepara el archivo, lo que realmente
#le da sus valores, o asigna es la madre de aca abajo
Wasabi = N_datos - 1000
for i in range(N_datos):
    datos_numericos[0,i] = float(datos[i+1][0])
    datos_numericos[1,i] = float(datos[i+1][1])
#al principio no entendia que vergas pasaba    
    datos_numericos[2,i] = float(datos[i+1][2])
    datos_numericos[3,i] = float(datos[i+1][3])
          
```

```python
figure(figsize=(8,6),dpi = 1000)
for i in range(1,4):
    plot(datos_numericos[i,:],label='$N_{max}=%1.0f$' %i)
legend(loc='lower left')
show() 
```
