```python
def deriv_h2(x,f):
    #Esto funciona si la malla es uniforme
    f_prima = zeros_like(f) # es un array con puros ceros LOL de ña longitud de f
    h = x[1]-x[0]
    for i in range(len(f)-1):
        f_prima[i] = (f[i+1]-f[i])
    f_prima[-1] = (f[-1]-f[-2])
    return f_prima/h
```

Comencemos con $f(x)$

```python
x = linspace(-5,5,100)
f = e**(-x**2)*sin(10*x)
f_prima = -e**(-x**2)*(2*x*sin(10*x)-10*cos(10*x))
f_prima_ord2 = deriv_h2(x,f)
plot(x,f,label='Analitica')
plot(x,f_prima_ord2,label='Orden 2')
legend()
show()
```
